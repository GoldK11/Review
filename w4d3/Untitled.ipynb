```python
from stack import *
```

```python
class Calculator:
    
    # 1. 설정 초기화 
    def __init__(self):
        self.orgexp = None
        self.fixexp = None
        self.result = None
    
    def set_orgexp(self,exp):
        self.orgexp = exp.replace(' ','')
        self.fixexp = None
        self.result = None
    
    def get_orgexp(self):
        return self.orgexp
    
    
    # 2. 편의함수
    # 2-1. 가중치 반환
    def get_weight(self,oprt):
        if oprt == '*' or oprt == '/':
            return 9
        elif oprt == '+' or oprt == '-':
            return 7
        elif oprt == '(':
            return 5
    
    # 2-2. 후위표기법 변환 (self.orgexp >> self.fixexp)
    # 식 리스트 exp_list , 연산자 스택  oprt_stack
    def convert_fix(self):
        exp_list = []
        oprt_stack = Stack()
            
        for ch in self.get_orgexp():
            if ch.isdigit():
                exp_list.append(ch)      
            else:
                if oprt_stack.empty() or ch=='(':
                    oprt_stack.push(ch)
                elif ch==')':
                    op = oprt_stack.pop()
                    while op != '(':
                        exp_list.append(op)
                        op = oprt_stack.pop()
                else:
                    if self.get_weight(ch) > self.get_weight(oprt_stack.peek()):
                        oprt_stack.push(ch)
                    else:
                        while oprt_stack and self.get_weight(ch) <= self.get_weight(oprt_stack.peek()):
                            op = oprt_stack.pop()
                            exp_list.append(op)
                        oprt_stack.push(ch)
        while oprt_stack:
            exp_list.append(oprt_stack.pop())       
        
        self.fixexp = ''.join(exp_list)                  
                    
    # 2-3. 연산 기능 정의
    def calc_oprds(self,oprd1,oprd2,oprt):
        if oprt == '+':
            return oprd1 + oprd2
        elif oprt == '-':
            return oprd1 - oprd2
        elif oprt == '*':
            return oprd1 * oprd2
        elif oprt == '/':
            return oprd1 // oprd2
            
    
    # 2-4. 식에 연산기능 도입
    # oprd_stack
    def calculate(self):
        oprd_stack = Stack()
        
        for ch in self.get_fixexp():
            if ch.isdigit():
                oprd_stack.push(int(ch))
            else:
                oprd2 = oprd_stack.pop()
                oprd1 = oprd_stack.pop()
                oprd_stack.push(self.calc_oprds(oprd1,oprd2,ch))
        
        self.result = oprd_stack.pop()
    
    
    # 3. user에게 공개되는 인터페이스
    # 3-1. 후위표기로 변환된 식 반환
    def get_fixexp(self):
        if not self.orgexp:
            return None
        if not self.fixexp:
            self.convert_fix()
        return self.fixexp
         
    #3-2. 연산 결과 반환
    def get_result(self):
        if not self.orgexp:
            return None
        if not self.result:
            self.calculate()
        return self.result  
```

```python
if __name__ == "__main__":
    calc = Calculator()
    while 1:
        exp = input("수식을 입력하세요(종료 : 0) : ")
        if exp == '0':
            break
        
        calc.set_orgexp(exp)
        print('후위표기로 변환: ',calc.get_fixexp())
        print('결과: ',calc.get_result())
                
```
